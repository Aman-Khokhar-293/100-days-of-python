```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
x = np.arange(1, 6)
y1 = [3, 5, 7, 9, 11]
y2 = [2, 4, 6, 8, 10]
```

```python
 # Line plot with styling
plt.plot(x, y1, label="Series 1", color="blue", marker="o", linestyle="--")
plt.plot(x, y2, label="Series 2", color="red", marker="s", linestyle="-.")

plt.title("Styled Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Pie chart
sizes = [30, 25, 20, 15, 10]
labels = ["A", "B", "C", "D", "E"]
plt.pie(sizes, labels=labels, autopct="%1.1f%%", startangle=140)
plt.title("Pie Chart Example")
plt.show()
```

```python
# Box plot
data = [np.random.normal(0, std, 100) for std in range(1, 4)]
plt.boxplot(data, vert=True, patch_artist=True)
plt.title("Box Plot Example")
plt.show()
```
