```python
import matplotlib.pyplot as plt
import numpy as np

# -----------------------------
# Line Plot
# -----------------------------
days = np.arange(1, 8)
temps = [30, 32, 34, 33, 31, 29, 28]

plt.plot(days, temps, marker='o', color='b', linestyle='--', label="Temperature")
plt.title("Daily Temperature")
plt.xlabel("Day")
plt.ylabel("Temperature (°C)")
plt.legend()
plt.show()
```

```python

# -----------------------------
# Bar Chart
# -----------------------------
products = ["A", "B", "C", "D"]
sales = [120, 150, 90, 200]

plt.bar(products, sales, color="orange")
plt.title("Product Sales")
plt.xlabel("Products")
plt.ylabel("Units Sold")
plt.show()
```

```python
# -----------------------------
# Scatter Plot
# -----------------------------
study_hours = [2, 3, 4, 5, 6, 7, 8]
marks = [50, 55, 60, 65, 70, 78, 85]

plt.scatter(study_hours, marks, color="green")
plt.title("Study Hours vs Marks")
plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.show()
```

```python

# -----------------------------
# Histogram
# -----------------------------
dice_rolls = np.random.randint(1, 7, 1000)
plt.hist(dice_rolls, bins=6, color="purple", edgecolor="black")
plt.title("Dice Roll Distribution")
plt.xlabel("Dice Face")
plt.ylabel("Frequency")
plt.show()
```
