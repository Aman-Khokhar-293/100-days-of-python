```python

# List Example – Grocery List
grocery_list = ["milk", "bread", "eggs"]
print("Original List:", grocery_list)

# Adding items
grocery_list.append("butter")        # Add at end
grocery_list.insert(1, "jam")        # Add at specific position
print("Updated List:", grocery_list)

# Removing items
grocery_list.remove("bread")
popped_item = grocery_list.pop()     # Removes last item
print("List after removal:", grocery_list)
print("Popped Item:", popped_item)

# Indexing and Slicing
print("First Item:", grocery_list[0])
print("Sliced List:", grocery_list[0:2])

# Nested List (2D list)
matrix = [[1, 2], [3, 4]]
print("Matrix Element [1][1]:", matrix[1][1])


# Tuple Example – Immutable collection
coordinates = (10.5, 20.8)
print("Tuple:", coordinates)
print("First Coordinate:", coordinates[0])

# Tuple with mixed data
student = ("Aman", 20, "B.Tech")
print("Student Info:", student)
```
