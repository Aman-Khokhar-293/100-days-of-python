```python
# Day 73 â€“ DBSCAN (Concept)

import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import make_moons
from sklearn.cluster import DBSCAN

# Generate a non-linear dataset (moons)
X, _ = make_moons(n_samples=300, noise=0.07, random_state=42)
```

```python

# ---------------------------
# Apply DBSCAN
# ---------------------------
dbscan = DBSCAN(eps=0.2, min_samples=5)  # eps=radius, min_samples=neighbors
y_pred = dbscan.fit_predict(X)
```

```python

# ---------------------------
# Visualization
# ---------------------------
plt.figure(figsize=(7, 5))
sns.scatterplot(x=X[:, 0], y=X[:, 1], hue=y_pred, palette="viridis", s=60)
plt.title("DBSCAN Clustering on Moons Dataset")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```
