```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Example dataset (Seaborn built-in)
tips = sns.load_dataset("tips")

# -----------------------------
# Basic Scatter Plot
# -----------------------------
sns.scatterplot(x="total_bill", y="tip", data=tips, hue="sex", style="time")
plt.title("Scatter Plot: Bill vs Tip")
plt.show()
```

```python
# -----------------------------
# Bar Plot
# -----------------------------
sns.barplot(x="day", y="total_bill", data=tips, ci="sd", palette="viridis")
plt.title("Average Bill per Day")
plt.show()
```

```python
# -----------------------------
# Count Plot
# -----------------------------
sns.countplot(x="day", data=tips, palette="Set2")
plt.title("Count of Records by Day")
plt.show()
```

```python
# -----------------------------
# Box Plot
# -----------------------------
sns.boxplot(x="day", y="total_bill", data=tips, palette="coolwarm")
plt.title("Box Plot: Bill Distribution by Day")
plt.show()
```

```python
# -----------------------------
# Violin Plot
# -----------------------------
sns.violinplot(x="day", y="total_bill", data=tips, palette="muted")
plt.title("Violin Plot: Bill Distribution by Day")
plt.show()
```

```python

# -----------------------------
# Pair Plot
# -----------------------------
sns.pairplot(tips[["total_bill", "tip", "size"]])
plt.suptitle("Pairplot of Tips Dataset", y=1.02)
plt.show()
```
