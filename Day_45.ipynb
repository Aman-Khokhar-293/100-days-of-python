```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load example dataset
students = sns.load_dataset("penguins")  # Using penguins dataset for demo

# -----------------------------
# Seaborn Themes
# -----------------------------
sns.set_theme(style="darkgrid")
sns.scatterplot(x="flipper_length_mm", y="body_mass_g", data=students, hue="species")
plt.title("Scatter Plot with Darkgrid Theme")
plt.show()

sns.set_theme(style="whitegrid")
sns.boxplot(x="species", y="body_mass_g", data=students, palette="Set2")
plt.title("Box Plot with Whitegrid Theme")
plt.show()
```

```python

# -----------------------------
# Color Palettes
# -----------------------------
sns.set_palette("coolwarm")
sns.histplot(students["body_mass_g"], kde=True)
plt.title("Distribution Plot with Coolwarm Palette")
plt.show()
```

```python

# -----------------------------
# Heatmap (Correlation)
# -----------------------------
corr = students.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap="YlGnBu", linewidths=0.5)
plt.title("Correlation Heatmap")
plt.show()
```

```python

# -----------------------------
# KDE Plot
# -----------------------------
sns.kdeplot(x="flipper_length_mm", y="body_mass_g", data=students, fill=True, cmap="mako")
plt.title("KDE Plot Example")
plt.show()
```
