```python
# Day 75 â€“ Dimensionality Reduction using PCA

import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA

# ---------------------------
# Load the Iris dataset
# ---------------------------
iris = load_iris()
X = pd.DataFrame(iris.data, columns=iris.feature_names)
y = iris.target
target_names = iris.target_names
```

```python
# ---------------------------
# Apply PCA (2 Components)
# ---------------------------
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

# Explained variance
print("Explained Variance Ratio:", pca.explained_variance_ratio_)
```

```python

# ---------------------------
# Visualization in 2D
# ---------------------------
plt.figure(figsize=(8, 6))
sns.scatterplot(x=X_pca[:, 0], y=X_pca[:, 1], hue=y, palette="viridis", s=70)
plt.title("PCA - 2D Visualization of Iris Dataset")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.show()
```

```python
# ---------------------------
# 3D Visualization (Optional)
# ---------------------------
from mpl_toolkits.mplot3d import Axes3D

pca_3d = PCA(n_components=3)
X_pca3 = pca_3d.fit_transform(X)

fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection="3d")
ax.scatter(X_pca3[:, 0], X_pca3[:, 1], X_pca3[:, 2], c=y, cmap="viridis", s=60)
ax.set_title("PCA - 3D Visualization of Iris Dataset")
ax.set_xlabel("PC1")
ax.set_ylabel("PC2")
ax.set_zlabel("PC3")
plt.show()
```
