```python
#1
print("for loop using range()")
for i in range(1,5):
  print("count ",i)

#2
print("Multiplication Table of 5 ")
for i in range(1,11):
  print(f"5 * {i} = {5*i}")

#3
print("Factorial of 5 Using While loop :")
num = 5
fact = 1
i = 1
while i <= num:
  fact = fact*i
  i=i+1
print("factorial of 5 : ",fact)

#4
print("count vowels in a 'Hello World' : ")
text = "Hello World"
vowels = "aeiouAEIOU"
count = 0
for char in text :
  if char in vowels:
    count +=1             # count = count +1
print("Total vowels : ",count)

#5
print("sum of digits in a number 1234:")
number = 1234
sum_digits = 0
while number > 0:
  digits = number % 10
  sum_digits += digits
  number //= 10
print("sum of digits : ",sum_digits)
```

```python
#6
print("Right-angle triangle pattern:")
rows = 5
for i in range(1, rows + 1):
    print("*" * i)

#7
print("Loop Control Statements:")

#8
print("Using break:")
for i in range(10):
    if i == 5:
        break
    print(i, end=" ")
print()

#9
print("Using continue:")
for i in range(5):
    if i == 2:
        continue
    print(i, end=" ")
print()

#10
print("Using pass:")
for i in range(3):
    pass                # pass means do nothing
print("Pass statement executed (nothing happened).")
```
