```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load Titanic dataset from Seaborn
titanic = sns.load_dataset("titanic")
```

```python

# -----------------------------
# Data Cleaning
# -----------------------------
print(titanic.isnull().sum())   # Check missing values
titanic = titanic.dropna(subset=["age", "embarked"])  # Drop missing rows
```

```python

# -----------------------------
# Basic Analysis
# -----------------------------
print(titanic["survived"].value_counts())
print(titanic.groupby("sex")["survived"].mean())
```

```python

# -----------------------------
# Visualization
# -----------------------------
# Survival count by gender
sns.countplot(x="sex", hue="survived", data=titanic, palette="Set2")
plt.title("Survival Count by Gender")
plt.show()
```

```python
# Survival by class
sns.barplot(x="class", y="survived", data=titanic, ci=None, palette="muted")
plt.title("Survival Rate by Class")
plt.show()
```

```python

# Age distribution by survival
sns.kdeplot(data=titanic, x="age", hue="survived", fill=True)
plt.title("Age Distribution by Survival")
plt.show()
```
