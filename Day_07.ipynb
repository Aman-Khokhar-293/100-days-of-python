```python
#Defining Functions using def
def greet():
    print("Hello, Welcome to Python Functions!")

greet()

# Function Parameters (positional, keyword, default)
def user_info(name, age=18):  # default parameter age=18
    print(f"Name: {name}, Age: {age}")

user_info("Aman")   # uses default age
user_info("Krish", 22)   # positional argument
user_info(age=25, name="Ravi")   # keyword arguments

# --- Return Statement ---
def square(num):
    return num * num

print("Square of 5:", square(5))

# Variable Scope (Local vs Global)
x = 10  # global variable

def modify():
    global x
    x = 20  # modifies global variable
    y = 5   # local variable
    print("Inside function - x:", x)
    print("Inside function - y:", y)

modify()
print("Outside function - x:", x)
```

```python
def add(a, b): return a + b
def subtract(a, b): return a - b
def multiply(a, b): return a * b
def divide(a, b): return a / b if b != 0 else "Cannot divide by zero"

print("Add:", add(10, 5))
print("Subtract:", subtract(10, 5))
print("Multiply:", multiply(10, 5))
print("Divide:", divide(10, 5))
```

```python
def factorial(n):
    if n == 0 or n == 1:
        return 1
    return n * factorial(n - 1)

num = 5
print(f"Factorial of {num} is {factorial(num)}")
```
