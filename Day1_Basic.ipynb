```python
# Variables & datatypes
name = "Aman"        # String type
age = 21             # Integer type
height = 5.8         # Float type
is_student = True     # bool
x,y,z = 10,20,30     # Multiple variable assigment
```

```python
# Swapping varriables
a = 5
b = 8
a,b = b,a     # Swapped values
print("a = ",a,"b = ",b)
```

```python
# Type casting examples

# String to int
num_str = "100"          # str
num_int = int(num_str)   # int
print(num_int + 50)

# Int to float
n = 7                    # int
f = float(n)             # float
print(f)

# Float to int
pi = 3.14                # float
approx = int(pi)         # int
print(approx)

# Int to string
a = 42                   # int
a_str = str(a)           # str
print("The answer is: " + a_str)
```

```python
# Input and output with formatting

# Taking input (default type is str)
name = input("Enter your name: ")      # str
age = int(input("Enter your age: "))   # int

# Output using print()
print("Hello", name + "!")
print("You are", age, "years old.")

# Using f-strings (formatted string)
print(f"Nice to meet you, {name}. You are {age} years old.")
```

```python
# Arithmetic operations

a = 15        # int
b = 4         # int

print("Addition:", a + b)
print("Subtraction:", a - b)
print("Multiplication:", a * b)
print("Division:", a / b)
print("Floor Division:", a // b)
print("Modulus:", a % b)
print("Exponentiation:", a ** b)
```

```python
# 1. Add two numbers entered by the user
num1 = int(input("Enter first number: "))   # int
num2 = int(input("Enter second number: "))  # int
print("Sum is:", num1 + num2)

# 2. Greet user using name and age
name = input("Enter your name: ")           # str
age = int(input("Enter your age: "))        # int
print(f"Hello {name}, you are {age} years old!")

# 3. Celsius to Fahrenheit converter
celsius = float(input("Enter temperature in Celsius: "))  # float
fahrenheit = (celsius * 9/5) + 32                         # float
print(f"Temperature in Fahrenheit: {fahrenheit}")

# 4. Simple Interest calculator
p = float(input("Enter Principal amount: "))  # float
r = float(input("Enter Rate of Interest: "))  # float
t = float(input("Enter Time (in years): "))   # float

si = (p * r * t) / 100                         # float
print(f"Simple Interest is: {si}")
```
