```python
#String Slicing
text = "Hello, Python!"
print("Original:", text)
print("First 5 chars:", text[:5])
print("Last 6 chars:", text[-6:])
print("Reversed:", text[::-1])

# Common String Methods
sample = "  python programming  "
print("Upper:", sample.upper())
print("Lower:", sample.lower())
print("Strip:", sample.strip())
print("Replace:", sample.replace("python", "java"))
print("Split:", sample.split())
print("Join:", "-".join(["Learn", "Python", "Fast"]))

# Searching in Strings
print("Find 'prog':", sample.find("prog"))
print("Count 'p':", sample.count("p"))
print("Starts with 'py':", sample.startswith("py"))
print("Ends with 'ing':", sample.endswith("ing"))

# String Formatting
name = "Aman"
age = 21
city = "Ahmedabad"
print(f"My name is {name}, I am {age} years old, from {city}.")
print("My name is {}, I am {} years old, from {}.".format(name, age, city))
print("My name is %s, I am %d years old, from %s." % (name, age, city))

```

```python
# ---------------- Mini Projects ----------------

# 1. Reverse a String
string = "Python"
print("\nReversed String:", string[::-1])

# 2. Palindrome Checker
word = "madam"
if word == word[::-1]:
    print(f"{word} is a Palindrome")
else:
    print(f"{word} is not a Palindrome")

# 3. Vowel Counter
sentence = "Learning Python is fun"
vowels = "aeiouAEIOU"
count = sum(1 for char in sentence if char in vowels)
print("Number of vowels:", count)

# 4. User Details Formatter
user_name = "Aman"
user_age = 21
user_profession = "Data Scientist"
print(f"User: {user_name} | Age: {user_age} | Profession: {user_profession}")
```
