```python
# Sets Example
my_set = {1, 2, 3, 4}
print("Original Set : ", my_set)

# Add an element to the set
my_set.add(5)
print("Set after adding an element : ", my_set)

# Removing elements\
my_set.remove(2)
print("After removing 2 :",my_set)

# Set operations
set_a = {1,2,3}
set_b = {3,5,6}
print("Union: ",set_a.union(set_b))
print("Intersection : ",set_a.intersection(set_b))
print("Difference : ",set_a.difference(set_b))
```

```python
# Dictionaries Example
my_dict = {"name":"Aman","age":21,"city":"Ahmadabad"}
print("Original Dictionaries : ",my_dict)

# Accessing values
print("Name:",my_dict["name"])

# Adding and updating
my_dict["profession"] = "Data Scientist"
my_dict.update({"age": 22})
print("Updated Dictionary:", my_dict)

# Dictionary methods
print("Keys:", my_dict.keys())
print("Values:", my_dict.values())
print("Items:", my_dict.items())
```

```python
items = [1, 2, 2, 3, 4, 4, 5]
unique_items = set(items)
print("Unique Items:", unique_items)
```

```python
# Dictionary: Student -> {Subject: Marks}
students = {
    "Aman": {"Math": 90, "Science": 85, "English": 78},
    "Yukti": {"Math": 88, "History": 92, "Science": 80},
    "Krish": {"English": 82, "Math": 95, "Geography": 89}
}

# Show each student's marks
for student, subjects in students.items():
    print(f"\n{student}'s Marks:")
    for subject, marks in subjects.items():
        print(f"  {subject}: {marks}")

# Collect all unique subjects using set
unique_subjects = set()
for subjects in students.values():
    unique_subjects.update(subjects.keys())

print("\nAll Unique Subjects:", unique_subjects)
```
