```python
# --- List Comprehension Examples ---

# 1. Square of numbers 1 to 10
squares = [x**2 for x in range(1, 11)]
print("Squares:", squares)

# 2. Even numbers from a list
nums = [1, 2, 3, 4, 5, 6, 7, 8]
evens = [x for x in nums if x % 2 == 0]
print("Even numbers:", evens)

# 3. Nested List Comprehension â€“ 2D list of multiplication table
table = [[i * j for j in range(1, 6)] for i in range(1, 6)]
print("Multiplication Table:", table)
```

```python
# --- Dictionary Comprehension Examples ---

# 4. Word lengths dictionary
words = ["python", "code", "awesome"]
lengths = {word: len(word) for word in words}
print("Word lengths:", lengths)

# 5. Square of numbers dictionary
square_dict = {x: x**2 for x in range(1, 6)}
print("Square dict:", square_dict)
```
